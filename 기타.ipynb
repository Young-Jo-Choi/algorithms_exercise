```python
# 1026번 보물
N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

# A의 가장 작은 값과 B의 가장 큰 값이 곱해지도록 구성
temp = []
for _ in range(N):
    A_min = min(A)
    B_max = max(B)
    temp.append(A_min*B_max)
    A.remove(A_min)
    B.remove(B_max)  
print(sum(temp))  
```
