```python
# 1026번 보물
N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

# A의 가장 작은 값과 B의 가장 큰 값이 곱해지도록 구성
temp = []
for _ in range(N):
    A_min = min(A)
    B_max = max(B)
    temp.append(A_min*B_max)
    A.remove(A_min)
    B.remove(B_max)  
print(sum(temp))  
```

```python
# 2436 공약수
import math
gcd, lcm = map(int,input().split())
if gcd==lcm:
    print(gcd, lcm)
else:
    minimum = lcm*2
    div = lcm//gcd  # div = num1'*num2' # num1 = num1'*gcd, num2=num2'*gcd, num1', num2'는 서로소
    for num1_d in range(1,div//2+1):
        if div % num1_d==0:
            num2_d = div//num1_d
            if math.gcd(num1_d,num2_d)==1:
                if num1_d+num2_d < minimum:
                    minimum = num1_d+num2_d
                    global_num1_d = num1_d
                    global_num2_d = num2_d
    print(global_num1_d*gcd, global_num2_d*gcd)
```
