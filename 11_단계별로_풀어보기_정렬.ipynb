# 수 정렬하기
## 문제
N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

## 입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000)이 주어진다. 둘째 줄부터 N개의 줄에는 수 주어진다. 이 수는 절댓값이 1,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.

## 출력
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

```python
# 거품정렬 사용 (O(log(n^2)))
# sort 함수는 너무 간단하므로 지양
N = int(input())
numbers = []
for i in range(N):
    numbers.append(int(input()))

no_change = True
cnt = 1
while cnt>0:
    cnt = 0
    for n in  range(0,len(numbers)-1):
        if numbers[n] > numbers[n+1]:
            big_number = numbers[n]
            small_number = numbers[n+1]
            numbers[n] = small_number
            numbers[n+1] = big_number
            cnt += 1

for n in numbers:
    print(n)
```

# 수 정렬하기2
## 문제
N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

## 입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000,000)이 주어진다. 둘째 줄부터 N개의 줄에는 수 주어진다. 이 수는 절댓값이 1,000,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.

## 출력
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

```python
# 다음 퀵정렬을 구상했으나 최악의 경우 O(n^2)이므로 시간초과 혹은 메모리 초과가 뜬다. 그러므로 다른 정렬 방법을 사용

import sys
def qsort(numbers):

    if len(numbers)<=1:
        return numbers
    standard = numbers[0]
    numbers.remove(standard)
    front = [number for number in numbers if number < standard]
    back = [number for number in numbers if number > standard]

    return qsort(front)+[standard]+qsort(back)

N = int(input())
nums = []

# 1,000,000까지 입력이 가능하니 느린 input 대신 다음 함수로 데이터 입력하도록 바꿈
for i in range(N):
    nums.append(int(sys.stdin.readline()))
sorted = qsort(nums)
for i in sorted:
    sys.stdout.write(str(i)+'\n')
```

```python
# 병합정렬
import sys
def merge(front,back):
    i = 0
    j = 0
    numbers_new = []
    while len(front) > i and len(back) > j:
        if front[i] < back[j]:
            numbers_new.append(front[i])
            i += 1
        else:
            numbers_new.append(back[j])
            j += 1
    if len(front)==i:
        # extend 사용시 시간초과 뜸
        while len(back) > j:
            numbers_new.append(back[j])
            j+=1
    elif len(back)==j:
        while len(front) > i:
            numbers_new.append(front[i])
            i+=1
    return numbers_new

def msort(numbers):
    if len(numbers)<=1:
        return numbers
    splitted = len(numbers)//2
    front = msort(numbers[:splitted])
    back = msort(numbers[splitted:])
    sorted_numbers = merge(front,back)
    return sorted_numbers

N = int(input())
nums = []
# 1,000,000까지 입력이 가능하니 느린 input 대신 다음 함수로 데이터 입력하도록 바꿈
for i in range(N):
    nums.append(int(sys.stdin.readline()))
sorted = msort(nums)
for i in sorted:
    sys.stdout.write(str(i)+'\n')
```

# 수 정렬하기3

```python
# 카운팅 정렬
# 실패 
import sys
def csort(nums):
    count = [0]*(max(nums)+1)
    for num in nums:
        count[num] += 1
    for i in range(1,len(count)):
        count[i] += count[i-1]
    result = [0]*(len(nums))
    for num in nums:
        idx = count[num]
        result[idx-1] = num
        count[num] -= 1
    return result

N = int(input())
nums = []
# 10,000,000까지 입력이 가능하니 느린 input 대신 다음 함수로 데이터 입력하도록 바꿈
for i in range(N):
    nums.append(int(sys.stdin.readline()))
sorted = csort(nums)
for i in sorted:
    sys.stdout.write(str(i)+'\n')
```

# 통계학
## 문제
수를 처리하는 것은 통계학에서 상당히 중요한 일이다. 통계학에서 N개의 수를 대표하는 기본 통계값에는 다음과 같은 것들이 있다. 단, N은 홀수라고 가정하자.

산술평균 : N개의 수들의 합을 N으로 나눈 값<br>
중앙값 : N개의 수들을 증가하는 순서로 나열했을 경우 그 중앙에 위치하는 값<br>
최빈값 : N개의 수들 중 가장 많이 나타나는 값<br>
범위 : N개의 수들 중 최댓값과 최솟값의 차이<br>
N개의 수가 주어졌을 때, 네 가지 기본 통계값을 구하는 프로그램을 작성하시오.

## 입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 500,000)이 주어진다. 단, N은 홀수이다. 그 다음 N개의 줄에는 정수들이 주어진다. 입력되는 정수의 절댓값은 4,000을 넘지 않는다.

## 출력
첫째 줄에는 산술평균을 출력한다. 소수점 이하 첫째 자리에서 반올림한 값을 출력한다.

둘째 줄에는 중앙값을 출력한다.

셋째 줄에는 최빈값을 출력한다. 여러 개 있을 때에는 최빈값 중 두 번째로 작은 값을 출력한다.

넷째 줄에는 범위를 출력한다.

```python
import sys
from collections import Counter
def mode(sorted_numbers:list)->int:
    if len(sorted_numbers) == 1:
        return sorted_numbers[0]
    countings = Counter(sorted_numbers).most_common()
    if (countings[0][1] == countings[1][1]) and (len(countings)>1):
        return countings[1][0]
    else:
        return countings[0][0]

N = int(input())

for i in range(N):
    numbers.append(int(sys.stdin.readline()))

print(round(sum(numbers)/len(numbers)))
sorted_numbers = sorted(numbers)
print(sorted_numbers[len(numbers)//2])
print(mode(sorted_numbers))
print(max(numbers)-min(numbers))
```

# 소트인사이드
## 문제
배열을 정렬하는 것은 쉽다. 수가 주어지면, 그 수의 각 자리수를 내림차순으로 정렬해보자.

첫째 줄에 정렬하려고 하는 수 N이 주어진다. N은 1,000,000,000보다 작거나 같은 자연수이다.

# 좌표 정렬하기
## 문제
2차원 평면 위의 점 N개가 주어진다. 좌표를 x좌표가 증가하는 순으로, x좌표가 같으면 y좌표가 증가하는 순서로 정렬한 다음 출력하는 프로그램을 작성하시오.

## 입력
첫째 줄에 점의 개수 N (1 ≤ N ≤ 100,000)이 주어진다. 둘째 줄부터 N개의 줄에는 i번점의 위치 xi와 yi가 주어진다. (-100,000 ≤ xi, yi ≤ 100,000) 좌표는 항상 정수이고, 위치가 같은 두 점은 없다.

## 출력
첫째 줄부터 N개의 줄에 점을 정렬한 결과를 출력한다.

```python
import sys
N = int(input())
xy_list = []
for _ in range(N):
    xy = tuple(map(int,sys.stdin.readline().split()))
    xy_list.append(xy)
xy_list = sorted(xy_list, key=lambda x: x[1])
xy_list = sorted(xy_list, key=lambda x: x[0])
for x,y in xy_list:
    print(f'{x} {y}')
```

# 단어 정렬
## 문제
알파벳 소문자로 이루어진 N개의 단어가 들어오면 아래와 같은 조건에 따라 정렬하는 프로그램을 작성하시오.

1. 길이가 짧은 것부터
2. 길이가 같으면 사전 순으로

## 입력
첫째 줄에 단어의 개수 N이 주어진다. (1 ≤ N ≤ 20,000) 둘째 줄부터 N개의 줄에 걸쳐 알파벳 소문자로 이루어진 단어가 한 줄에 하나씩 주어진다. 주어지는 문자열의 길이는 50을 넘지 않는다.

## 출력
조건에 따라 정렬하여 단어들을 출력한다. 단, 같은 단어가 여러 번 입력된 경우에는 한 번씩만 출력한다.

```python
import sys
N = int(input())
words = []
for _ in range(N):
    words.append(input())

words_zip = list(set(map(lambda x:(len(x),x), words)))
words_sorted = sorted(words_zip, key=lambda x:x[1])
words_sorted = sorted(words_sorted, key=lambda x:x[0])
# print로 출력하니까 출력형식이 잘못되었다는 에러가 나옴(이유는 모름)
for n,x in words_sorted:
    sys.stdout.write(x+'\n')
```

# 나이순 정렬
## 문제
온라인 저지에 가입한 사람들의 나이와 이름이 가입한 순서대로 주어진다. 이때, 회원들을 나이가 증가하는 순으로, 나이가 같으면 먼저 가입한 사람이 앞에 오는 순서로 정렬하는 프로그램을 작성하시오.

## 입력
첫째 줄에 온라인 저지 회원의 수 N이 주어진다. (1 ≤ N ≤ 100,000)

둘째 줄부터 N개의 줄에는 각 회원의 나이와 이름이 공백으로 구분되어 주어진다. 나이는 1보다 크거나 같으며, 200보다 작거나 같은 정수이고, 이름은 알파벳 대소문자로 이루어져 있고, 길이가 100보다 작거나 같은 문자열이다. 입력은 가입한 순서로 주어진다.

## 출력
첫째 줄부터 총 N개의 줄에 걸쳐 온라인 저지 회원을 나이 순, 나이가 같으면 가입한 순으로 한 줄에 한 명씩 나이와 이름을 공백으로 구분해 출력한다.

```python
import sys
N = int(input())
members = []
for i in range(N):
    members.append(input().split() + [i])
members_zip = list(map(lambda x:(int(x[0]),x[1],x[2]),members))
member_sorted = sorted(members_zip, key=lambda x:x[2])
member_sorted = sorted(member_sorted, key=lambda x:x[0])
for age,name,_ in member_sorted:
    sys.stdout.write(str(age)+' '+name+'\n')
```

# 좌표 압축
## 문제
수직선 위에 N개의 좌표 X1, X2, ..., XN이 있다. 이 좌표에 좌표 압축을 적용하려고 한다.

Xi를 좌표 압축한 결과 X'i의 값은 Xi > Xj를 만족하는 서로 다른 좌표의 개수와 같아야 한다.

X1, X2, ..., XN에 좌표 압축을 적용한 결과 X'1, X'2, ..., X'N를 출력해보자.

## 입력
첫째 줄에 N이 주어진다.

둘째 줄에는 공백 한 칸으로 구분된 X1, X2, ..., XN이 주어진다.

## 출력
첫째 줄에 X'1, X'2, ..., X'N을 공백 한 칸으로 구분해서 출력한다.

## 제한
1 ≤ N ≤ 1,000,000
-109 ≤ Xi ≤ 109

```python
import sys
N = int(input())
numbers = list(map(int, sys.stdin.readline().split()))

number_set = set(numbers)
numbers_sorted = sorted(number_set)
numbers_zip = list(zip(numbers_sorted, range(len(numbers_sorted)),number_set))
number_sorted = sorted(numbers_zip, key=lambda x:x[2])
number_dict = dict(map(lambda x:(x[0],x[1]),number_sorted))

for num in numbers:
    sys.stdout.write(str(number_dict[num]) + ' ')
```
