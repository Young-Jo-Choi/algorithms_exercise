# 수 정렬하기
## 문제
N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

## 입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000)이 주어진다. 둘째 줄부터 N개의 줄에는 수 주어진다. 이 수는 절댓값이 1,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.

## 출력
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

```python
# 거품정렬 사용 (O(log(n^2)))
# sort 함수는 너무 간단하므로 지양
N = int(input())
numbers = []
for i in range(N):
    numbers.append(int(input()))

no_change = True
cnt = 1
while cnt>0:
    cnt = 0
    for n in  range(0,len(numbers)-1):
        if numbers[n] > numbers[n+1]:
            big_number = numbers[n]
            small_number = numbers[n+1]
            numbers[n] = small_number
            numbers[n+1] = big_number
            cnt += 1

for n in numbers:
    print(n)
```

# 수 정렬하기2
## 문제
N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

## 입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000,000)이 주어진다. 둘째 줄부터 N개의 줄에는 수 주어진다. 이 수는 절댓값이 1,000,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.

## 출력
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

```python
# 다음 퀵정렬을 구상했으나 최악의 경우 O(n^2)이므로 시간초과 혹은 메모리 초과가 뜬다. 그러므로 다른 정렬 방법을 사용

import sys
def qsort(numbers):

    if len(numbers)<=1:
        return numbers
    standard = numbers[0]
    numbers.remove(standard)
    front = [number for number in numbers if number < standard]
    back = [number for number in numbers if number > standard]

    return qsort(front)+[standard]+qsort(back)

N = int(input())
nums = []

# 1,000,000까지 입력이 가능하니 느린 input 대신 다음 함수로 데이터 입력하도록 바꿈
for i in range(N):
    nums.append(int(sys.stdin.readline()))
sorted = qsort(nums)
for i in sorted:
    sys.stdout.write(str(i)+'\n')
```

```python
# 병합정렬
import sys
def merge(front,back):
    i = 0
    j = 0
    numbers_new = []
    while len(front) > i and len(back) > j:
        if front[i] < back[j]:
            numbers_new.append(front[i])
            i += 1
        else:
            numbers_new.append(back[j])
            j += 1
    if len(front)==i:
        # extend 사용시 시간초과 뜸
        while len(back) > j:
            numbers_new.append(back[j])
            j+=1
    elif len(back)==j:
        while len(front) > i:
            numbers_new.append(front[i])
            i+=1
    return numbers_new

def msort(numbers):
    if len(numbers)<=1:
        return numbers
    splitted = len(numbers)//2
    front = msort(numbers[:splitted])
    back = msort(numbers[splitted:])
    sorted_numbers = merge(front,back)
    return sorted_numbers

N = int(input())
nums = []
# 1,000,000까지 입력이 가능하니 느린 input 대신 다음 함수로 데이터 입력하도록 바꿈
for i in range(N):
    nums.append(int(sys.stdin.readline()))
sorted = msort(nums)
for i in sorted:
    sys.stdout.write(str(i)+'\n')
```
